```python
#merge means to join 2 or more dataset
#suppose create first dictionary
d1={"name":["renu","kamal","rohini"],"maths":[34,45,34]}
d2={"name":["renu","kamal","rohini","pankaj"],"science":[46,36,32,30]}
```

```python
print(d1)
```

```python
print(d2)
```

```python
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
```

```python
print(df1)
```

```python
print(df2)
```

```python
#merge both dataframe df1 and df2 , merge() inbuilt function of pandas
#Equal join means Equal merge
pd.merge(df1,df2,on="name")
```

```python
#outer merge
pd.merge(df1,df2,on="name",how="outer") #NaN means null values
```

```python
#outer merge
result=pd.merge(df1,df2,on="name",how="outer") #NaN means null values
print(result)
```

```python
#To check null values in dataframe result
result.isnull().sum()
```

```python
result["maths"].fillna(result["maths"].mean(),inplace=True) 
#fill the marks of maths of last student of mean value of maths column
#permanently means inplace=True
```

```python
result
```

```python
#to change the datatype of maths column from float to int
result["maths"]=result["maths"].astype("int")
```

```python
result["maths"]
```

```python
#concatenate function
#create first dictionary :
d1={'name':['rekah','suresh','deepak'],'age':[23,21,25]}
#create second dictionary
d2={'city':['Bhopal','Indore','Delhi'],
    'mobile':[9827652345,9876543212,9876547897]}
```

```python
print(df2)
```

```python
''' name  age  city  mobile'''
#concat() means to join 2 data frame column wise means axis=1
df3=pd.concat([df1,df2],axis=1)
print(df3)
```
